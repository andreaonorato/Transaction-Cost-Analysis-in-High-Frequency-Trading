SPY ETF data <br>
File: SPY_1min_sample_firstratedata.zip <br>
Entries: 207.824 <br>
Time: 2022 - 1 min data <br>
Columns: timestamp    open    high     low   close  volume

```python
import zipfile
import pandas as pd

# Path to your ZIP file
zip_path = 'data/SPY_1min_sample_firstratedata.zip'

# Open the ZIP file
with zipfile.ZipFile(zip_path, 'r') as z:
    # List all files in the ZIP archive
    file_list = z.namelist()
    print("Files in ZIP archive:", file_list)
    
    # Assuming the file of interest is the first one in the ZIP archive
    target_file = file_list[0]  # Adjust index if needed
    
    # Open the target file as a pandas DataFrame
    with z.open(target_file) as f:
        df = pd.read_csv(f)
        # Display the first few lines of the DataFrame
        print(df.head())
print(df.shape)
```

High-frequency Chinese market dataset <br>
File: minDataSum.zip <br>
Entries: Different size for each stock <br>
Time: From 2013 to 2019 depending on the stock - 1 min data <br>
Columns: Time      Open      High       Low     Close  Volume Amount  Open_Interest

```python
import zipfile
import pandas as pd

# Path to your ZIP file
zip_path = 'data/minDataSum.zip'

# Open the ZIP file
with zipfile.ZipFile(zip_path, 'r') as z:
    # List all files in the ZIP archive
    file_list = z.namelist()
    
    for file in file_list:
        print(f"Processing file: {file}")
        
        # Open each file in the ZIP archive
        with z.open(file) as f:
            # Check if the file is a CSV (to avoid processing non-CSV files)
            if file.endswith('.csv'):
                df = pd.read_csv(f)
                
                # Print the file name, location (path), first 5 rows, and shape
                print(f"File name: {file}")
                print(f"Location (path): {zip_path}/{file}")
                print("First 5 rows of the DataFrame:\n", df.head())
                print("Shape of the DataFrame:", df.shape)
                print("\n" + "="*50 + "\n")
```
